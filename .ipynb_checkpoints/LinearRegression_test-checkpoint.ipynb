```python
import numpy as np
```

```python
from LinearRegression import LinearRegression
```

```python
X = np.random.rand(1000, 3)
```

```python
y = 2*X[:, 0] - 4*X[:, 1] + 3*X[:, 2] + 0.5
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y)
```

```python
lin_reg.weights
```

```python
lin_reg.bias
```

```python
lin_reg.predict(X)
```

```python
lin_reg.evaluate(X, y)
```
